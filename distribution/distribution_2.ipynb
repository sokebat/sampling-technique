```python
import pandas as pd
import numpy as np

data = {
    "E": [
        12, 15, 18, 20, 22, 24, 25, 27, 28, 30,
        32, 33, 35, 36, 38, 40, 42, 43, 45, 46,
        48, 50, 52, 53, 55, 56, 58, 60, 62, 63,-50, -20, 160, 180,
        65, 66, 68, 70, 72, 73, 75, 76, 78, 80,
        82, 83, 85, 86, 88, 90, 92, 93, 95, 98
    ]
}

df = pd.DataFrame(data)


```

```python
lst = df["E"].tolist()
```

## Maximum And Minimum value

```python
max_val = df['E'].max()
min_val = df['E'].min()

print("max value is:", max_val)
print("min value is:", min_val)
```

## mean

```python
mean = df["E"].sum()/len(lst)

print(f"The Mean E is {mean}")
```

## Variance

```python
sum = 0
for i in lst:
  sum += (i-mean)**2
variance = sum/(len(lst)-1)

print("variance : ", variance)
```

## Standard Deviation

```python
sd = variance** 0.5

print("standard Deviation ",sd)
```

## 1st Standard Deviation ,2nd  Standard Deviation, 3rd Standard Deviation

```python
first_standard = mean+ 1*(sd)
second_standard = mean+ 2*(sd)
third_standard = mean+ 3*(sd)

print(f"The First Standard Deviation is {first_standard}")
print(f"The Second Standard Deviation is {second_standard}")
print(f"The Third Standard Deviation is {third_standard}")
```

## Calculating the Quartiles Q1,Q2,Q3

```python
last = lst[-1]
first = lst[0]

# print(last)
# print(first)

pos_q1 = len(lst)/4
pos_q2 = len(lst)/2
pos_q3 = 3*len(lst)/4

print(pos_q1)
print(pos_q2)
print(pos_q3)


```

```python

q1  = lst[13-1]
q2 = lst[27-1]
q3 = lst[40-1]

print(f"The First Quartile is {q1}")
print(f"The Second Quartile is {q2}")
print(f"The Third Quartile is {q3}")
```

##  InterQuartiles Range's

```python
IQR = q3 - q1
print(f"The Inter Quartiles Range is {IQR}")
```

## Lower Outliers and Upper Outlie

```python
low_outliers = q1-1.5*IQR
high_outliers = q3+1.5*IQR

print(f"The Lower Outliers are {low_outliers}")
print(f"The Higher Outliers are {high_outliers}")
```

Based on the dataset E, the lower outlier boundary is -22.0 and the upper outlier boundary is 130.

```python

### Chekcing the OUtliers ::
count = 0
for i in lst:
  if i >= high_outliers or i <= low_outliers:
    count += 1
  ### The Total Number of the Outliers
print(f"The total number of the OUtliers is {count}")
```

## 1st Central Moment

```python
first_sum = 0
for i in lst:
    first_sum += (i - mean)**1

central_moment1 = first_sum / len(lst)
print("1st Central Moment:", central_moment1)
```

## 2nd Central Moment

```python
second_sum  = 0
for i in lst:
  second_sum += (i-mean)**2
cental_moment2 = second_sum/(len(lst)-1) ## Sample Variance
print(cental_moment2)
```

## 3rd Central Moment

```python
third_sum  = 0
for i in lst:
  third_sum += (i-mean)**3
cental_moment3 = third_sum/(len(lst)-1) ## Sample Variance
print(cental_moment3)
```

## 4th Cental Moment


```python
fourth_sum = 0
for i in lst:
  fourth_sum += (i-mean)**4
cental_moment4 = fourth_sum/(len(lst)-1)
print(cental_moment4)
```

# Conclusion, From the Given Data Set, It is clear that

### The 2nd central moment (≈ 1367.8) represents the variance of the dataset, showing a very wide dispersion around the mean. This high variability is mainly caused by the extreme values −50, −20, 160, and 180, which significantly increase the spread.


### The 3rd central moment (≈ 23558.78) is positive, indicating a positively skewed distribution. This confirms that the right tail is longer, primarily influenced by the large observations 160 and 180.


### The 4th central moment (≈ 10223093.78) relates to kurtosis and is considerably large, confirming a heavy-tailed distribution. This suggests a higher probability of extreme values compared to a normal distribution, driven by outliers on both the lower and upper ends.

##Skewenes

```python
if cental_moment3 == 0:
  print("Symmetric")
elif cental_moment3 > 0:
  print("Right Skewed")
else:
  print("Left Skewed")
```

The distribution of the data is right skewed (positively skewed). This means that most of the values are concentrated on the left side, with a long tail stretching to the right.

##Kurtotis

```python

# Kurtosis
beta2 = cental_moment4 / (cental_moment2**2)

print("Kurtosis (β2):", beta2)
```

```python
# Kurtosis type
if beta2 > 3:
    print("Distribution: Leptokurtic")
elif beta2 < 3:
    print("Distribution: Platykurtic")
else:
    print("Distribution: Mesokurtic")
```

The distribution is leptokurtic, which means it has a sharper peak and heavier tails than a normal distribution.
