```python
import pandas as pd

data = {
    "A": [12, 15, 14, 16, 18, 17, 19, 20, 21, 22, 23, 24, 25, 26, 27],
    "B": [35, 34, 36, 38, 40, 39, 41, 42, 43, 44, 45, 46, 47, 48, 49],
    "C": [102, 98, 100, 105, 110, 108, 112, 115, 118, 120, 122, 125, 128, 130, 132],
    "D": [5.2, 5.5, 5.4, 5.6, 5.8, 5.7, 5.9, 6.0, 6.1, 6.2, 6.3, 6.4, 6.5, 6.6, 6.7],
    "E": [200, 210, 205, 215, 220, 218, 225, 230, 235, 240, 245, 250, 255, 260, 265]
}

df = pd.DataFrame(data)
df
```

##Range

```python
A_range = df["A"].max() - df["A"].min()


print(f"Range of A: {A_range}")
```

##Variance:

```python
A = df["A"].to_list()

mean_A = sum(A) / len(A)
print(f"Mean of A: {mean_A}")

sum_A = 0
for x in A:
    sum_A += (x - mean_A) ** 2

print(f"Sum of squared deviations (A): {sum_A}")

variance_A = sum_A / (len(A) - 1)
print(f"Sample Variance of A: {variance_A}")
```

##Standard Devation

```python
sd_A = variance_A ** 0.5
print(f"Standard Deviation (A): {sd_A}")
```

##Z Score

```python
value = 16
z_A = (value - mean_A) / sd_A
print(f"Z-score (A): {z_A}")
```

##Conclusion:

Range is 15, mean is 19.93, variance is 21.06 and standard deviation is 4.58.
